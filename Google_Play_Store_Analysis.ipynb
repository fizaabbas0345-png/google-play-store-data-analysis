```python
import numpy as np
df['Size'] = df['Size'].replace('Varies with device', np.nan)
```

```python
# convert KB to MB
def convert_size(x):
    if pd.isna(x):
        return np.nan

    if 'M' in str(x):
        return float(x.replace('M', ''))

    if 'k' in str(x):
        return float(x.replace('k', '')) / 1024

    return np.nan
```

```python
df['Size_MB'] = df['Size'].apply(convert_size)
```

```python
df['Size_MB'].head()
```

```python
df['Size_MB'] = df['Size_MB'].fillna(df['Size_MB'].median())
```

```python
import matplotlib.pyplot as plt
```

```python
# Which categories contain the most apps?


df['Category'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 App Category')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=45)
plt.show()
```

```python
# How are app ratings distributed?

df['Rating'].hist(bins=20)
plt.title('Distribution of Apps Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# top rated apps
top_apps = df.nlargest(10, 'Rating')

plt.figure(figsize=(10,5))
plt.bar(top_apps['App'], top_apps['Rating'])

plt.xticks(rotation=90)
plt.title('Top Rated Apps')
plt.show()
```

```python
# What percentage of apps are free versus paid?
df['Type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Free vs Paid Apps')
plt.ylabel('')

plt.show()
```

```python
# Which app categories attract the most downloads?
top_installs = (
    df.groupby('Category')['Installs']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_installs.plot(kind='bar')

plt.title('Top Categories by Total Installs')
plt.xlabel('Category')
plt.ylabel('Total Installs')
plt.xticks(rotation=90)

plt.show()
```

```python
# Is there a relationship between reviews and installs?
plt.scatter(df['Reviews'], df['Installs'])

plt.title('Reviews vs Installs')
plt.xlabel('Reviews')
plt.ylabel('Installs')

plt.show()
```

```python
top_rated = (
    df.groupby('Category')['Rating']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

top_rated.plot(kind='bar')

plt.title('Top Rated Categories')
plt.xlabel('Category')
plt.ylabel('Average Rating')

plt.show()
```
